**1.6 That means jpg has less channels then png so before applying GAN lets convert png to jpg. I wanted to assign the dataset path as I am using it frequently.**

```python
DATASET_PATH = 'pokemons'

def png_to_jpg(image_path):
    images = []
    fill_color = (255,255,255)

    for img in os.listdir(image_path):
        im = Image.open(image_path+'/{}'.format(img))
        if img.split('.')[1] == 'png':
            im = im.convert("RGBA")
            if im.mode in ('RGBA', 'LA'):
                bg = Image.new(im.mode[:-1], im.size, fill_color)
                bg.paste(im, im.split()[-1]) # omit transparency
                im = bg 
        images.append(np.array(im))
    return images
```

```python
CONVERTED_DATASET_IMAGES = png_to_jpg(DATASET_PATH)
```

# **2. GAN Genaration and Discrimination Phase to Produce new color pokemon images**

**2.1 Import necessary Libraries for Image Generation**

**2.2 Define the necessary parameters needed further in one place**

```python
# Generation resolution (64 X 64) - square 
# Training data is also have to scale to this.

# Note GENERATE_RES 4 or higher blows Google CoLab's memory (tested before) and have not been tested extensivly.

# Generation resolution factor (1=32, 2=64, 3=96, 4=128, etc.)

GENERATE_RES = 2 

GENERATE_SQUARE = 32 * GENERATE_RES # hieght/width of the images

#RGB channel for color image produce
IMAGE_CHANNELS = 3

# window configuration for previewing images 
PREVIEW_ROWS = 4
PREVIEW_COLS = 7
PREVIEW_MARGIN = 16

# Size vector to generate images from
# Noise vector size which will be used for generation
SEED_SIZE = 100 

# Model parameters Configuration
#FINAL_DATASET = CONVERTED_DATASET_IMAGES
EPOCHS = 10000
BATCH_SIZE = 32
BUFFER_SIZE = 60000

print(f"I will generate {GENERATE_SQUARE}px square images.")
```

**2.3 Rescale and Preprocess the images**

```python
def preprocess(image_path):  
    training_data = []
    for filename in tqdm(os.listdir(image_path)):
        path = os.path.join(image_path,filename)
        image = Image.open(path).resize((GENERATE_SQUARE,GENERATE_SQUARE),Image.ANTIALIAS).convert('RGBA')
        image.load() # required for png.split()
        background = Image.new("RGBA", image.size, (255, 255, 255))
        image = Image.alpha_composite(background, image).convert('RGB')
        training_data.append(np.asarray(image))
    training_data = np.reshape(training_data,(-1,GENERATE_SQUARE,GENERATE_SQUARE,IMAGE_CHANNELS))
    training_data = training_data.astype(np.float32)
    training_data = training_data / 127.5 - 1.
    return training_data
```

```python
training_data = preprocess(DATASET_PATH)
```

```python
#image
```

**2.4 Batch and shuffle Data**

```python
train_dataset = tf.data.Dataset.from_tensor_slices(training_data).shuffle(BUFFER_SIZE).batch(BATCH_SIZE)
```

**2.5 Pokemon Image Generation GAN network Design**

Parameters Selection:

https://machinelearningmastery.com/how-to-train-stable-generative-adversarial-networks/

```python
def pokemon_generator(seed_size, channels):
    model = Sequential()

    model.add(Dense(4*4*256,activation="relu",input_dim=seed_size))
    model.add(Reshape((4,4,256)))

    model.add(UpSampling2D())
    model.add(Conv2D(256,kernel_size=4,padding="same"))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Activation("relu"))

    model.add(UpSampling2D())
    model.add(Conv2D(256,kernel_size=4,padding="same"))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Activation("relu"))
   
    # Output resolution, additional upsampling
    model.add(UpSampling2D())
    model.add(Conv2D(128,kernel_size=4,padding="same"))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Activation("relu"))

    model.add(UpSampling2D(size=(GENERATE_RES,GENERATE_RES)))
    model.add(Conv2D(128,kernel_size=4,padding="same"))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Activation("relu"))

    # Final CNN layer
    model.add(Conv2D(channels,kernel_size=4,padding="same"))
    model.add(Activation("tanh"))

    return model
```

**2.6 Discrimination Network Model**

```python
def pokemon_discriminator(image_shape):
    model = Sequential()

    model.add(Conv2D(64, kernel_size=4, strides=2, input_shape=image_shape, padding="same"))
    model.add(LeakyReLU(alpha=0.2))

    model.add(Dropout(0.25))
    model.add(Conv2D(128, kernel_size=4, strides=2, padding="same"))
    model.add(ZeroPadding2D(padding=((0,1),(0,1))))
    model.add(BatchNormalization(momentum=0.8))
    model.add(LeakyReLU(alpha=0.2))

    model.add(Dropout(0.25))
    model.add(Conv2D(256, kernel_size=4, strides=2, padding="same"))
    model.add(BatchNormalization(momentum=0.8))
    model.add(LeakyReLU(alpha=0.2))

    model.add(Dropout(0.25))
    model.add(Conv2D(512, kernel_size=4, strides=2, padding="same"))
    model.add(BatchNormalization(momentum=0.8))
    model.add(LeakyReLU(alpha=0.2))

    model.add(Dropout(0.25))
    model.add(Flatten())
    model.add(Dense(1, activation='sigmoid'))

    return model
```

**2.7 Define and write a function to save the outputs**

```python
def save_images(noise,epoch):
    image_array = np.full(( 
      PREVIEW_MARGIN + (PREVIEW_ROWS * (GENERATE_SQUARE+PREVIEW_MARGIN)), 
      PREVIEW_MARGIN + (PREVIEW_COLS * (GENERATE_SQUARE+PREVIEW_MARGIN)), 3), 
      255, dtype=np.uint8)

    generated_images = generator.predict(noise)
    
    generated_images_for_save = generated_images
    generated_images_for_save *= 255

    if not os.path.exists('generated_images'):
        os.mkdir('generated_images')
    
    if epoch < EPOCHS:
        img = tf.keras.preprocessing.image.array_to_img(generated_images_for_save[1])
        img.save("generated_images/generated_img_%d_%d.png" % (epoch, 1))
    elif epoch == EPOCHS:
        for i in range(4):
            img = tf.keras.preprocessing.image.array_to_img(generated_images_for_save[i])
            img.save("generated_images/generated_img_%d_%d.png" % (epoch, i))

    # Rescale images 0 - 1
    generated_images = 0.5 * generated_images + 0.5

    image_count = 0
    for row in range(PREVIEW_ROWS):
        for col in range(PREVIEW_COLS):
            r = row * (GENERATE_SQUARE+16) + PREVIEW_MARGIN
            c = col * (GENERATE_SQUARE+16) + PREVIEW_MARGIN
            image_array[r:r+GENERATE_SQUARE,c:c+GENERATE_SQUARE] = generated_images[image_count] * 255
            image_count += 1

    im = Image.fromarray(image_array)
    
    return im
```

**2.9 Functions and parameters for calculating the losses**

```python
# This method returns a helper function to compute cross entropy loss
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)

def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    total_loss = real_loss + fake_loss
    return total_loss

def generator_loss(fake_output):
    return cross_entropy(tf.ones_like(fake_output), fake_output)
```

**2.10 Define the optimizers (Adam is used here)**

```python
generator_optimizer = tf.keras.optimizers.Adam(1.5e-6,0.5)
discriminator_optimizer = tf.keras.optimizers.Adam(5.0e-6,0.5)
```

**2.11 One by One image batch training and calculating the losses for generation and dicrimination.**

```python
# Notice the use of `tf.function`
# This annotation causes the function to be "compiled".
@tf.function
def train_step(images):
    seed = tf.random.normal([BATCH_SIZE, SEED_SIZE])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        generated_images = generator(seed, training=True)

        real_output = discriminator(images, training=True)
        fake_output = discriminator(generated_images, training=True)

        gen_loss = generator_loss(fake_output)
        disc_loss = discriminator_loss(real_output, fake_output)


        gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
        gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

        generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
        discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))
    
    return gen_loss,disc_loss
```

**2.12 The function to combined the training to produce the final model saved and claculate the average losses**

```python
def train(dataset, epochs):
    fixed_seed = np.random.normal(0, 1, (PREVIEW_ROWS * PREVIEW_COLS, SEED_SIZE))

    for epoch in range(epochs):
        #epoch_start = time.time()

        gen_loss_list = []
        disc_loss_list = []

        for image_batch in dataset:
            t = train_step(image_batch)
            gen_loss_list.append(t[0])
            disc_loss_list.append(t[1])

        g_loss = sum(gen_loss_list) / len(gen_loss_list)
        d_loss = sum(disc_loss_list) / len(disc_loss_list)

        
        print (f'Epoch {epoch+1}: gen_loss={g_loss}, disc_loss={d_loss}')
        #I want to save images after each 1000 epochs.
        if np.floor((epoch+1)/1000)==(epoch+1)/1000:
            generated_sample = save_images(fixed_seed,epoch)
        if np.floor((epoch)/1000)==0:
            generated_sample = save_images(fixed_seed,epoch)
    #generator.save('/kaggle/working/generator_model.h5')
    generator.save('generator_model.h5')
    
    return generated_sample
```

**2.13 Finally, run the complete process with the number of epochs and the preprocessed dataset.**

```python
%%time
# image of generations
generated_sample = train(train_dataset, EPOCHS)
```
